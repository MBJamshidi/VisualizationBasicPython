# Margins

Set or retrieve autoscaling margins.
The padding added to each limit of the axes is the margin times the data interval. All input parameters must be floats within the range [0, 1]. Passing both positional and keyword arguments is invalid and will raise a TypeError. If no arguments (positional or otherwise) are provided, the current margins will remain in place and simply be returned.
Specifying any margin changes only the autoscaling; for example, if xmargin is not None, then xmargin times the X data interval will be added to each end of that interval before it is used in autoscaling.

```python

#Margins

Cities=['Kermanshah', 'Eslam Abad', 'Sahnneh', 'Paveh', 'Javanroud', 'Ravansar', 'Tehran', 'Mashahd', 'Songhor']
Income=[1137, 123.5, 3133.6, 3616, 1462, 168, 720, 5519.5, 267.8]
plt.figure(figsize=(10, 9), dpi=80)
#c=color
Colors_income=['green', 'blue', 'pink', 'yellow', 'brown', 'black', 'orange', 'tomato', 'violet']
#s=size
Size_income=np.array([1137, 123.5, 3133.6, 3616, 1462, 168, 720, 5519.5, 267.8])
#scatter
plt.scatter(np.arange(9), Income, c=Colors_income, s=Size_income)

#Margine
plt.margins(0.1)



#plt.xticks([0,1,2,3,4,5,6,7,8], ['Kermanshah', 'Eslam Abad', 'Sahnneh', 'Paveh', 'Javanroud', 'Ravansar', 'Tehran', 'Mashahd', 'Songhor'])
plt.show()
```

# Multiple Plot, Line Style=ls'-', Marker='+' , Size Line=mew, Legend 

```python
Years=[2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010]
Inflation1=[11, 12.5, 13.6, 14, 16, 14, 12, 18, 20, 19.5, 27.8]
Inflation2=[4, 7.5, 5, 6, 8.3, 3.1, 5.2, 4, 4.8, 7, 4.8]
plt.title("Iinflation-1 Vs Inflation-2")
plt.plot(Years, Inflation1, ls='-', marker='+', mew=20)
plt.plot(Years, Inflation2, ls='--', marker='*', mew=10)
plt.legend(['Inflation 1', 'Inflation 2'], loc='best')
plt.show()
```

# Annotate and Grid

```python
Years=[2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010]
Inflation1=[11, 12.5, 13.6, 14, 16, 14, 12, 18, 20, 19.5, 27.8]
Inflation2=[4, 7.5, 5, 6, 8.3, 3.1, 5.2, 4, 4.8, 7, 4.8]
plt.title("Iinflation-1 Vs Inflation-2")
plt.plot(Years, Inflation1, ls='-', marker='+', mew=20)
plt.plot(Years, Inflation2, ls='--', marker='*', mew=10)
plt.legend(['Inflation 1', 'Inflation 2'], loc='best')


#GRID
plt.grid()

#annotate
plt.annotate("2004 Economic Crisis", xytext=(2002, 22), fontsize=10, xy=(2004, 16) , arrowprops=dict(facecolor='silver', width=4) ) 
#xy text: Position of the xy text

#annotate
plt.annotate("2004 Economic Crisis", xytext=(2004, 27), fontsize=14, xy=(2010, 27.8) , arrowprops=dict(facecolor='green', width=8) ) 
#xy text: Position of the xy text


#annotate
plt.annotate("Migration Crisis", xytext=(2004, 7), fontsize=12, xy=(2006, 5.2) , arrowprops=dict(facecolor='blue', width=6) ) 
#xy text: Position of the xy text

plt.show()
```

# Strip Plot

A strip is basically a scatter plot where the x axis represents a categorical variable.

# Swarm Plot

A swarm plot is very similar to a strip plot, yet the locations of points are adjusted automatically to avoid overlap even if the jitter value is not applied.

# BoxPlot

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.
Input data can be passed in a variety of formats, including:
Vectors of data represented as lists, numpy arrays, or pandas Series objects passed directly to the x, y, and/or hue parameters.
A “long-form” DataFrame, in which case the x, y, and hue variables will determine how the data are plotted.
A “wide-form” DataFrame, such that each numeric column will be plotted.
An array or list of vectors.

```python
sb.boxplot(x='OP', y='JAM1', data=DataFrame1)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sb.boxplot(x='MODEL', y='JAM1', data=DataFrame1)
plt.show()
```

```python
sb.boxplot(x='OP', y='JAM3', data=DataFrame1)
plt.show()
```

# Pair Plot

A pairs plot allows us to see both distribution of single variables and relationships between two variables . Pair plots are a great method to identify trends for follow-up analysis and, fortunately, are easily implemented in Python!
seaborn.pairplot
seaborn.pairplot(data, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None, kind='scatter', diag_kind='auto', markers=None, height=2.5, aspect=1, corner=False, dropna=True, plot_kws=None, diag_kws=None, grid_kws=None, size=None)
Plot pairwise relationships in a dataset.
By default, this function will create a grid of Axes such that each numeric variable in data will by shared in the y-axis across a single row and in the x-axis across a single column. The diagonal Axes are treated differently, drawing a plot to show the univariate distribution of the data for the variable in that column.

```python
sb.pairplot(DataFrame1, hue="OP", palette="hls")
plt.show()
```

```python
sb.pairplot(DataFrame1, hue="OP", palette="husl")
plt.show()
```

```python
#Size of the points
sb.pairplot(DataFrame1, hue="OP", palette="hls", plot_kws={'s':100})
plt.show()
```

```python
sb.pairplot(DataFrame1, hue="MODEL", palette="hls", plot_kws={'s':100})
plt.show()
```
